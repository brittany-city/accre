```python
def convert_time(time):
   d = 86400
   h = 3600
   m = 60
   times = time.split(':')
   times = times[0].split('-') + [times[1]] + [times[2]]
   if len(times) == 3:
       times.insert(0,0)
   times = [int(t) for t in times]
   seconds = (times[0] * d) + (times[1] * h) + (times[2] * m) + times[3]
   return seconds
```

```python
accre_df['REQTIME'] = accre_df['REQTIME'].apply(convert_time)
accre_df['USEDTIME'] = accre_df['USEDTIME'].apply(convert_time)
```

```python
accre_df['CPUS_per_NODES']=accre_df['CPUS']/accre_df['NODES']
```

```python
def mem_type(mem):
   """identify whether the memory request is per node or per core"""
   if mem[-2:] == 'Mn':
       return 'node'
   elif mem[-1:] == 'M':
       return 'node'
   elif mem[-2:] =='Mc':
        return 'core'
```

```python
accre_df['REQMEM_TYPE'] = accre_df['REQMEM'].apply(mem_type)
```

```python
accre_df['USEDMEM_TYPE'] = accre_df['USEDMEM'].apply(mem_type)
```
