#### Job time converted to total seconds.

```python
accre_df.REQTIME.str.split('-')
```

```python
def convert_time(time):
   d = 86400
   h = 3600
   m = 60
   times = time.split(':')
   times = times[0].split('-') + [times[1]] + [times[2]]
   if len(times) == 3:
       times.insert(0,0)
   times = [int(t) for t in times]
   seconds = (times[0] * d) + (times[1] * h) + (times[2] * m) + times[3]
   return seconds
```

```python
accre_df['REQTIME'] = accre_df['REQTIME'].apply(convert_time)
accre_df['USEDTIME'] = accre_df['USEDTIME'].apply(convert_time)
accre_df['REMAIN'] = accre_df['REQTIME'] - accre_df['USEDTIME']
```

#### Convert memory to Megabyters per core.

```python
accre_df['CPUS_per_NODES']=accre_df['CPUS']/accre_df['NODES']
```

```python
accre_df['REQMEM'] = accre_df['REQMEM'].str.rstrip('Mnc').astype('int64')
```

```python
accre_df['REQMEM'] = accre_df['REQMEM']/accre_df['CPUS_per_NODES']
```

```python
accre_df['USEDMEM']=accre_df['USEDMEM'].str.rstrip('M').astype('float')
```

```python
accre_df['USEDMEM'] = accre_df['USEDMEM']/accre_df['CPUS_per_NODES']
```

```python
accre_df['SUBID']=accre_df['JOBID'].str.split('_').str[1]
accre_df['JOBID']=accre_df['JOBID'].str.split('_').str[0]
```

```python
accre_df = accre_df.reindex(columns=['SUBMIT','EXITCODE','JOBID','SUBID','ACCOUNT','USER','REQMEM','USEDMEM','REQTIME','USEDTIME',])
```

```python
def run_result(code):
   """identify whether the run result was successful or failed"""
   if code == '0:0':
       return 'success'
   else:
     return 'failure'
```

```python
accre_df['EXITCODE']=accre_df['EXITCODE'].apply(run_result)
```

```python
accre_df['SUBMIT'] = pd.to_datetime(accre_df['SUBMIT'])
```
